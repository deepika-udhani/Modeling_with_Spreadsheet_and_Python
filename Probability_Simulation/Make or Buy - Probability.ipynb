```python
import numpy as np
import numpy_financial as npf
```

```python
Price = 40
BuyCost = 18
MakeCost = 7
Discrate = 0.15
MachCost = 400000
TaxRate = 0.4
```

```python
np_d=np.array([0,0,0,0,0])
np_cfb=np.array([0,0,0,0,0])
np_cfm=np.array([0,0,0,0,0])
np_deprec=np.array([0,100000,100000,100000,100000])
NPVtable=np.arange(1,1001)
```

```python
for y in range(0,1000):
    np_d[1]=np.random.triangular(10000, 14000, 16000)
    for x in range (2,5):
        np_d[x]=np_d[x-1]*(np.random.triangular(1.1, 1.2, 1.26))
    np_cfb=(np_d*(Price-BuyCost))*(1-TaxRate)
    np_cfm=(np_d*(Price-MakeCost)-np_deprec)*(1-TaxRate)+np_deprec
    npvbuy=npf.npv(Discrate, np_cfb)
    npvmake=npf.npv(Discrate, np_cfm)-MachCost
    npvnet=npvmake-npvbuy
    NPVtable[y]=npvnet
```

```python
print("NPV nets are ", NPVtable )
```

```python
AvgNPVdiff=np.mean(NPVtable)
print("Average NPV diff = ", AvgNPVdiff)
```

```python
makecount=0
for i in range(0,999):
    if NPVtable[i]>0:
        makecount=makecount+1
ProbMake=makecount/1000
print(ProbMake)
```
